```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import monai
from monai.transforms import LoadImage
from tifffile import imread
from monai.data import WSIReader
```

```python
tiff_dir = "../data/train/*"
tiff_file_paths = glob.glob(tiff_dir)
```

```python
print(len(tiff_file_paths))
```

```python
# TIFFファイルのパスを指定します。
file_path = tiff_file_paths[0]
print(file_path)
```

```python
# MONAIのLoadImage変換を使用して、画像データを読み込みます。
loader = LoadImage(reader=WSIReader, backend="cuCIM", dtype=float)
image = loader(file_path)[0].permute(1, 2, 0).numpy()
print(image.min(), image.max())
print(image.shape)
```

```python
plt.imshow(image/255)
plt.show()
```
